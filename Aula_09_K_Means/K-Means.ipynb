```python
# Importando as bibliotecas
import pandas as pd
```

```python
# Lendo o arquivo CSV
base = pd.read_csv('compras_supermercado.csv')
```

```python
# Verificando a quantidade de linhas e colunas
base.shape
```

```python
# Visualizando as 5 primeiras linhas
base.head()
```

```python
# Importando as bibliotecas
from sklearn.cluster import KMeans
```

```python
# Criando o modelo de KMeans
kmeans = KMeans(n_clusters=4, max_iter=1000)
```

```python
# Treinando o modelo com os dados
kmeans.fit(base)
```

```python
# Verificando os rótulos atribuídos a cada ponto de dados
kmeans.labels_
```

```python
# Verificando os centros dos clusters
kmeans.cluster_centers_
```

```python
# Verificando os centros dos clusters
pd.DataFrame(kmeans.cluster_centers_)
```

```python
# Verificando os centros dos clusters com os nomes das colunas
pd.DataFrame(kmeans.cluster_centers_, columns=base.columns)
```

```python
# Importando a biblioteca
import numpy as np
```

```python
# Criando um DataFrame com os rótulos e a frequência de cada cluster
grupo, frequencia = np.unique(kmeans.labels_, return_counts=True)
```

```python
# Visualizando os rótulos e a frequência
print(grupo)
print(frequencia)
```
