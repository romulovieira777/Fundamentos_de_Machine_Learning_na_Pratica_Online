```python
# Importando as bibliotecas
import pandas as pd
```

```python
# Lendo o arquivo xlsx
base = pd.read_excel("funcionarios.xlsx")
```

```python
# Visualizando as 5 primeiras linhas
base.head()
```

```python
# Removenos os valores nulos
del base['Unnamed: 0']
```

```python
# Verificando os dados
base
```

```python
# Selecionando as colunas de interesse
X = base.iloc[:, :-1]
y = base.iloc[:, -1]
```

```python
# Verificando os dados
X
```

```python
# Verificando os dados
y
```

```python
# Importando a biblioteca de árvore de decisão
from sklearn.tree import DecisionTreeClassifier
```

```python
# Criando o modelo
tree = DecisionTreeClassifier()
```

```python
# Treinando o modelo
tree.fit(X, y)
```

```python
# Amostrando a árvore de decisão
amostra = base.sample()
```

```python
# Verificando a amostra
amostra
```

```python
# Fazendo a previsão
tree.predict(amostra.iloc[:, :-1])
```

```python
# Importando a biblioteca para plotar a árvore
from sklearn.tree import export_graphviz
```

```python
# Criando o arquivo .dot
arquivo = open("arvore.dot", 'w')
```

```python
# Imprimindo a árvore
X.columns
```

```python
# Verificando os dados
base.linguagem.unique()
```

```python
# Exportando a árvore
export_graphviz(tree, out_file=arquivo, feature_names=X.columns, class_names=base.linguagem.unique())
```

```python
# Fechando o arquivo
arquivo.close()
```
