```python
# Carregando o dataset
base = pd.read_excel("funcionarios.xlsx")
```

```python
# Removendo a coluna "Unnamed: 0"
del base["Unnamed: 0"]
```

```python
# Visualizando as 5 primeiras linhas
base.head()
```

```python
# Selecionando as colunas de interesse
X = base.iloc[:, :-1]
```

```python
# Visualizando as linhas e colunas
X
```

```python
# Selecionando a última coluna
Y = base.iloc[:, -1]
```

```python
# Visualizando as linhas e colunas
Y
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Criando o classificador
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Treinando o classificador
knn.fit(X, Y)
```

```python
# Pegando uma amostra aleatória
amostra = base.sample()
```

```python
# Visualizando a amostra
amostra
```

```python
# Selecionando as colunas de interesse
amostra.iloc[:, :-1]
```

```python
# Pegando uma amostra e preparando para a previsão
knn.predict(amostra.iloc[:, :-1])
```
